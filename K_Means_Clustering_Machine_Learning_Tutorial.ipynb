```python
X = df # Feature vector and target variable

y = df['status_type']
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

X['status_type'] = le.fit_transform(X['status_type'])

y = le.transform(y)
```

```python
X.info()
```

```python
X.head()
```

```python
cols = X.columns
```

```python
from sklearn.preprocessing import MinMaxScaler

ms = MinMaxScaler()

X = ms.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns=[cols])
```

```python
X.head()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=0)

kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.inertia_
```

```python
labels = kmeans.labels_

# check how many of the samples were correctly labeled
correct_labels = sum(y == labels)

print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))
```

```python
print('Accuracy score: {0:0.2f}'. format(correct_labels/float(y.size)))
```

```python
from sklearn.cluster import KMeans
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)

kmeans.fit(X)

# check how many of the samples were correctly labeled
labels = kmeans.labels_

correct_labels = sum(y == labels)
print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))
print('Accuracy score: {0:0.2f}'. format(correct_labels/float(y.size)))
```

```python
from sklearn.decomposition import PCA

# Fit KMeans with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)

# Get cluster labels
labels = kmeans.labels_

# Get the PCA components
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

# Create a DataFrame for plotting
plot_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
plot_df['Cluster'] = labels

# Plot clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=plot_df, palette='Set1', legend='full')
plt.title('K-Means Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score
import matplotlib.pyplot as plt

# Calculate ARI for different numbers of clusters
ari_scores = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans.fit(X)
    labels = kmeans.labels_
    ari = adjusted_rand_score(y, labels)  # Replace y_true with true cluster labels if available
    ari_scores.append(ari)

# Plot ARI scores
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), ari_scores, marker='o')
plt.title('Adjusted Rand Index vs. Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Adjusted Rand Index')
plt.xticks(range(2, 11))
plt.grid(True)
plt.show()
```
